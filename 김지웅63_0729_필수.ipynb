Flavanoids와 Alcohol: Flavanoids와 Alcohol 간에 중간 정도의 양의 상관관계(0.236)가 있어서, Flavanoids 함량이 증가하면 Alcohol도 증가할 수 있다.

이러한 상관관계를 바탕으로 몇 가지 가설을 세울 수 있다.

가설 1: Total Phenols 함량이 높은 와인은 Flavanoids 함량도 높을 것이다. 이 관계를 더 자세히 알아보고, Total Phenols 수준에 따라 Flavanoids 수준이 유의미하게 차이나는지 확인할 수 있다.


- 귀무가설 (H0): Total Phenols 함량이 높은 와인과 낮은 와인의 Flavanoids 함량에는 유의미한 차이가 없다.
- 대립가설 (H1): Total Phenols 함량이 높은 와인은 Flavanoids 함량도 높다.
