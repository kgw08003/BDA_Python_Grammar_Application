공분산(covariance)는 두 변수 간의 관계를 나타내는 통계적 개념으로 두 변수가 함게 어떻게 변하는지를 측정하여 그 관계의 강도와 방향을 파악하는데 사용된다.


#### 공분산 구하는 공식

- 공분산(Cov(X, Y)) = Σ((X - μX) * (Y - μY)) / (n - 1)
- Cov(X, Y): X와 Y의 공분산
- X, Y: 각각의 변수
- μX, μY: X와 Y의 평균
- Σ: 합계를 나타내는 기호
- n: 데이터의 개수

##### 공분산이 양수인 경우
- X와 Y는 양의 관계를 가지며, 함께 증가 또는 감소한다,

##### 공분산이 음수인 경우
- X와 Y 사이에는 선형적인 관계가 없다. 하지만 다른 종류 관계가 있을 수 있으므로 주의가 필요하다.


공분산은 두 변수간의 선형적인 관계를 측정하기 때문에 데이터의 분산과 스케일에 영향을 받을 수 있는 것이다.
- 공분산 값만으로는 관계의 강도를 정확하게 파악하는 것은 매우 어려운 일이다.

이를 보안하기 위해서는 상관계수(Correlation Coeffecient)가 사용되며, 이는 공분산을 각 변수의 표준편차로 나눈 값을 의미한다,
- 상관관계는 -1과 1 사이의 값으로 정규화되며, 두 변수 간의 선형적인 관계의 강도와 방향을 더 정확하게 나타내는데 도움을 준다.
