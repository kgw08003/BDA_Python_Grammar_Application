### 타이타닉 데이터 컬럼 설명
- survivaed: 생존 여부 (0 = 사망, 1 = 생존)
- pclass : 객실의 등급 (1 = 1st, 2 = 2nd, 3 = 3rd)
- sex : 성별
- age: 나이
- sibsp : 동승한 형제자매, 배우자의 수
- parch : 동슨한 부모, 자녀 수
- fare : 운임 요금
- embarked : 탑승지 (C=Cherbourg, Q=Queenstown, S=Southampton)
- class : 객실의 등급 
- who : 남자, 여자 아이
- adult_male : 성인 남성인지
- deck : 선실 번호의 첫 알파벳
- embark_town : 탑승지 
- alive : 생존유무
- alone : 혼자 탑승했는지 (True는 혼자 탑승, False는 혼자 아님)

```python
tt.query("survived == 1").survived.count()
# 생존한 사람들의 수
```

```python
tt.query("survived == 0").survived.count()
# 사망한 사람들의 수
```

```python
tt.groupby('sex').agg(count=("survived",'sum'))
# 존한 사람들 중 남성과 여성의 수
```

```python
tt.groupby('pclass').agg(survive=('survived',sum))
# 객실 등급별 생존율
```

```python
tt.query('survived==1').age.mean()
# 생존자들의 평균연령
```

```python
tt.query('survived==1').groupby('embarked').agg(survive=('survived','count'))
# 탑승지별 생존자수
```

```python
tt['age'].mean()
# 전체 평균 나이
```

```python
tt.query("age >= 29.69911764705882 and alive == 'yes'") 
# 평균나이 이상 이고, 살아남은 인원이 포함한 전체
```

```python
tt.query("survived == 1")[['age','fare']].mean()
# 생존한 사람들의 평균 나이 및 티켓가격
```

```python
tt[['pclass','sex']]
# 객실 등급과 성별 나열
```

```python
tt[['pclass','sex']].sort_values('pclass')
# 객실등급과 성별 나열 (객실등급으로 오름차순)
```

```python
tt[['pclass','sex']].sort_values('pclass', ascending=False)
# 객실등급과 성별 나열 (객실등급으로 내림차순)
```

```python
tt[['pclass','sex']].sort_values(['pclass','sex'], ascending=[True,False])
# 객실등급과 성별 나열 (객실등급 및 성별으로 내림차순/ 둘다 겹치는 경우 우선순위는 객실등급)
```
