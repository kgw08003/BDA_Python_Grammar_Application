## ML 학습

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

- 결정트리를 사용하여 모델 학습 후 그리드 서치 사용하여 하이퍼 파라미터를 서치

```python
# train, test 데이터 분리
data = df1.drop('satisfaction', axis=1)
target = df1['satisfaction']
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=111)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

print("학습 데이터 점수: {}".format(accuracy_score(y_train, train_pred)))
print("평가 데이터 점수: {}".format(accuracy_score(y_test, test_pred)))
```

```python
model.get_params()
```

- 그리드 서치를 사용해서 모델의 성능을 향상

```python
# 그리드서치
params = {
    'max_depth': [4, 5, 6, 7, 8, None],
    'min_samples_split': [2, 3, 4, 5],
    'min_samples_leaf': [2, 3, 4, 5]
}

grid = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=params, cv=3)
grid.fit(X_train, y_train)
```

```python
# 최적의 estimator
grid.best_estimator_.get_params()
```

```python
grid_est = grid.best_estimator_
pred = grid_est.predict(X_test)

print("학습 데이터 점수: {}".format(accuracy_score(y_train, train_pred)))
print("평가 데이터 점수: {}".format(accuracy_score(y_test, pred)))
```

```python
grid_est = grid.best_estimator_
pred = grid_est.predict(X_test)

print("학습 데이터 점수: {}".format(accuracy_score(y_train, train_pred)))
print("평가 데이터 점수: {}".format(accuracy_score(y_test, pred)))
```

 - TensorFlow와 Keras를 사용하여 신경망 모델을 생성

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import accuracy_score

# 신경망 모델 생성
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# 모델 컴파일
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 모델 학습
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
# 학습 데이터 예측
train_pred_nn = (model.predict(X_train) > 0.5).astype(int)
train_accuracy_nn = accuracy_score(y_train, train_pred_nn)

# 테스트 데이터 예측
test_pred_nn = (model.predict(X_test) > 0.5).astype(int)
test_accuracy_nn = accuracy_score(y_test, test_pred_nn)
```

```python
print("신경망 - 학습 데이터 정확도: {:f}".format(train_accuracy_nn))
print("신경망 - 테스트 데이터 정확도: {:f}".format(test_accuracy_nn))
```
