- mpg객체에서 manufacturer, model, year 속성을 가져와서 i, j, k라는 빈 리스트에 각각 추가하는 작업

```python
i = []
j = []
k = []

for a,b,c in zip(mpg.manufacturer, mpg.model, mpg.year):
    i.append(a)
    j.append(b)
    k.append(c)
    
print(i)
```

```python
# 리스트 컴프리헨션 이용
i = [a for a, _, _ in zip(mpg.manufacturer, mpg.model, mpg.year)]
j = [b for _, b, _ in zip(mpg.manufacturer, mpg.model, mpg.year)]
k = [c for _, _, c in zip(mpg.manufacturer, mpg.model, mpg.year)]

print(i)
```

- mpg데이터 중 manufacturer ,year 그 중 year 2003.5 이상인 값들만 불러옴

```python
for a, b in zip(mpg.manufacturer,mpg.year):
    if b>=2003.5:
        print(a,b)
```

```python
# 리스트 컴프리헨션 이용
data = [(a, b) for a, b in zip(mpg.manufacturer, mpg.year) if b >= 2003.5]
for a, b in data:
    print(a, b)
```

- m5만 추출하는 함수(수업중 배운버전)

```python
def get_trans_spe(series): 
    tr_cnt =len([ _ for x in series if x[0] in 'm5'])
    return tr_cnt
```

```python
get_trans_spe(mpg.trans)
```

```python
mpg.groupby('manufacturer').agg({'trans':get_trans_spe}).reset_index()
```

- m5만 추출하는 함수 = len다음값에 _없이 구현한 버전

```python
def get_trans_spe2(series):
    tr_cnt = len([x for x in series if x[0] in 'm5'])
    return tr_cnt
```

```python
get_trans_spe2(mpg.trans)
```

```python
mpg.groupby('manufacturer').agg({'trans':get_trans_spe2}).reset_index()
```

```python
# trans가 auto(s6)인 경우 trans와 good 출력, auto(I5)인 경우 trans와 normal 출력, 그렇지 않은 경우 bad 출력
```

```python
for df_tr in mpg['trans']:
    if df_tr == 'auto(s6)':
        print(df_tr, 'good')
    elif df_tr == 'auto(l5)':
        print(df_tr, 'normal')
    else:
        print(df_tr, 'bad')
```

- category가 midsize인 자료들을 제조사별로 분류하기

```python
def get_cat_midsize(series):
    hwy_category = len([_ for x in series if x[0] in 'midsize'])
    return hwy_category
```

```python
get_cat_midsize(mpg.category)
```

```python
mpg.groupby('manufacturer').agg({'category':get_cat_midsize}).reset_index()
```
