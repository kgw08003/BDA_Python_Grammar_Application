### 데이터
- 데이터를 통계적으로 생각해 보자
- 질적 변수와 양적 변수
- 질적 변수 : 만족도, 혈액형 설문조사식의 변수들이다.
- 양적 변수 : 수치 시험 점수, 혈압, 양을 표현하는 것들
- 이런 변수들 -> 머신러닝을 위해서 인코딩, 임베딩 -> 변수를 수치화 하는 작업

> 척도

- 명의 척도, 순서 척도, 간격 척도, 비례 척도, 단순히 분류하기 위한 변수 = 성별, 전화번호 이런 값을 계산하면 의미는 없다. 전화번호 + 전화번호 의미 없다.
- 순서 척도 = 순서 관계, 대소 관계 설문조사 만족도, 0!5 점수, 성적 순위 1~50등 순서간의 관계를 보면 1등과 2등 차이와 10등과 11등 차이가 같다고 볼 수는 없다.
- 간격 척도 = 대소 관계인데 차이또한 의미가 있다. 온도 100도로 가면 더 높은 것 0도 낮은 것 60도 30도가 2배차이라고 말할 수는 없다.
- 비례 척도 = 대소 관계, 차이, 비 모두 의미가 있다. 길이 50cm 200cm 4배 길다.
- 간격 척도, 비례 척도 비교시 절대 0점 이야기를 한다. 온도 같은 경우는 0도가 있다. 길이는 0cm는 없다고 본다. 그 차이로 간격과 비례를 이해해도 된다.

- 척도에 따라서 통계분석은 달라지게 된다. 명목형, 순서형에 따라서 t-test, 카이제곱, 분산분석을 해야 하는지 이런 것들이 결국 데이터 타입에 따라 다 달라진다.

> 변수

- 이산형 변수, 연속형 변수
- 이산형 0,1,2,3,4 하나하나 값을 취하는 변수 서로 인접한 값의 사이 값은 존재하지 않는다. 0, 1, 2, 3
- 연속형 연속적인 값을 취한다. 0, 1 이더라도 그 안에 무수한 값이 있다. 0,1,0,001
- 이산형은 머신러닝에서 분류라는 모델링을 진행한 것
- 정답이 0,1,2 이진,다중분류의 문제를 해결하는 것
--------------------------------------------------------------------------
- 연속형은 머신러닝에서 회귀라는 모델링을 진행한 것 
- 연속적인 숫자를 예측하는 회귀로 문제를 해결하는 것 (연속적인 값)

- 이산형과 연속형의 스케일링 방법도 다 다르다. 
- 연속적인 값은 스케일링 가능하다 (정규화)

- 이산형은 사실 스케일 방법이 다르다. 문자형데이터인 경우에는 -> 인코딩할 수 있다. 임베딩할 수 있다.

- 표준편차, 분산 
- 데이터에 대한 산포도의 개념 
- 얼마나 데이터가 퍼져 있는가?

- A반은 전체 50점 20명 ( 평균 50 )
- B반은 10명 0점 0명 100점 10명 (평균 50)

- 산포도를 알기 위한 편차 
- 편차는 결국 각 데이터가 평균으로부터 어느정도 떨어져 있는가 이다.
- 영어 42점 평균 60 이 값은 평균으로 부터 18점 떨어져 있다.

```python
en_mean= np.mean(df.english) # english 열 평균 계산 en_mean = np.mean(df["english"])
```

```python
deviation=df.english- en_mean # 열의 각 값들의 평균값으로부터 편차 계산해서 df 에 새로운시리즈 또는 열로 저장
```

```python
np.sum(df.english- en_mean) # english" 열의 값들과 평균(en_mean)의 차이를 계산하고, 그 차이들의 총합을 구하는 코드
```

- 산포도 평균으로부터 떨어져 있는 차이를 나타낸 분산은 결국 평균이 항상 0에 가깝기 때문에
- 이걸 가지고 판단 하기가 어렵다.

- 표준편차를 사용하는 이유 
- 분산으로만 보기에는 어렵고, 데이터의 특성이 다 다를 것이고 동일한 지표를 사용해야 하기 때문에
- 산포도에서 지표로 보는 분산에 제곱근을 취한 표준편차를 사용하는 것 


```python
df['deviation'] = deviation
```

-표준편차는 산포도 지표에서 분산에 제곱근을 취한 것

```python
np.sqrt(np.var(df['english'])) #표준편차 #sqrt 제곱근
```

```python
np.std(df['english']) #표준편차 #std 바로 표준편차
```

### 데이터의 정규화
- 통일된 지표가 피룡한 것
- 다양한 변수들로 인한 지표값이 다 다르기 대문에 맞춰주는 것
- 정규화는 머신러닝에도, 추측통계에서도 사용하기 때문에 필수적으로 알아야 한다!

```python
## 표준화
# z-score 
z = (df.english - np.mean(df.english))/np.std(df.english)
```

```python
df['z-score'] = z
```
