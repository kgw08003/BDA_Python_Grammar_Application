<h2>확률</h2> : 하나의 사건이 일어날 수 있는 가능성을 수로 나타낸 것
- 같은 조건 아래에서 실험이나 관찰을 여러 번 반복할 때, 어떤 사건이 일어나는 상대도수가 일정한 값에 가까워지면 이 일정한 값을 그 사건이 일어날 확률이라고 한다.

<h3>확률의 성질</h3>
 
------------------------

- 어떤 사건이 일어날 확률을 p라 하면  0 ≤ p ≤ 1 dlek.
- 반듯이 일어나는 사건의 확률은 1이다.
- 절대로 일어나지 않는 사건의 확률은 0이다.

-------------

- 0과 1 사이의 값을 가질때 : 확률은 항상 0과 1 사이의 값을 가지며 여기서 0은 사건이 발생하지 않을 확률을 나타내며, 1은 사건이 반드시 발생할 확률을 나타낸다.

- 모든 가능한 결과들의 확률의 합은 1 이다. 어떤 사건의 가능한 결과들이 여러 개 라면, 각 결과의 확률을 모두 합하면 1이 된다. 이를 확률의 정규화 조건이라고 한다.

- 여사건의 확률: 어떤 사건 A의 확률을 P(A)라고 할 때, 사건 A가 발생하지 않을 확률을 P(A')라고 표현핟나. P(A')는 1 - p(A)로 계산된다.

- 합사건의 확률: 두 개 이상의 사건이 동시에 발생하는 경우를 합사건이라고 한다. 두 사건 A와 B의 합사건의 확률은 P(A∪B)로 표현되며, 이는 P(A) + P(B) - P(A∩B)로 계산되어 진다. 여기서 P(A∩B)는 사건 A와 B가 동시에 발생할 확률을 의미합니다.

- 조건부 확률: 어떤 사건 B가 발생했다는 가정하에 사건 A가 발생할 확률을 조건부 확률 P(A|B)로 표현한다. 조건부 확률은 P(A∩B) / P(B)로 계산되며. 이는 사건 B가 발생한 상황에서 사건 A가 발생할 확률을 나타낸다.

<h3>확률분포의 정의</h3>

---------------



- 확률분포는 확률 변수가 특정한 값을 가질 확률을 나타내는 함수를 의미한다.
- 확률 분포는 확률 변수의 종류에 따라 이산확률분포와 연속확률분포로 나눠진다.(쉽게말해서 셀 수 있는지 없는지에 따라 나뉜다.)

##### 이산확률 분포
- 이산확률변수의 확률분포를 의미
- 이산확률변수는 확률변수가 가질 수 있는 값의 개수를 셀 수 있다는 의미

##### 확률질량 함수
- 이산확률변수에서 특정 값에 대한 확률을 나타내는 함수이다.

##### 연속확률 분포
- 연속확률변수의 확률분포를 의미
- 연속확률변수는 확률변수가 가질 수 있는 값의 개수를 셀 수 없다는 의미

##### 확률밀도 함수
- 연속확률변수가 특정 구간에 포함될 확률을 의미
- 이산확률분포에서 확률 질량함수에 대응된다

##### 누적분포함수
- 주어진 확률 변수가 특정 값 보다 작거나 같은 확률을 나타내는 함수

##### 독립항등분포
- 두 개 이상의 확률변수를 고려할 때, 각 변수들이 통계적으로 독립이고, 동일한 확률분포를 가지고 있을때 독립항등분포를 따른다.
- iid라고 줄여서 부른다.

### 확률분포의 예시
- 베르누이분포 (Bernoulli Distribution):
- 베르누이분포는 이진 결과를 모델링하는 확률분포입다. 주사위를 던졌을 때 특정 숫자가 나오는 경우와 같이 두 가지 결과 중 하나가 나타날 때 사용됩다. 확률변수 X의 확률분포는 아래와 같다: P[X=k] = p^k * (1-p)^(1-k), 여기서 k는 0 또는 1이고, p는 성공(1)의 확률이다.

- 이항분포 (Binomial Distribution):
- 이항분포는 베르누이 시행을 여러 번 반복한 결과를 모델링하는 확률분포이다. 각 시행에서 성공할 확률이 p인 경우, 시행 횟수 n과 성공 횟수 k를 변수로 가지며, 확률변수 X의 확률분포는 아래와 같다: P[X=k] = (n C k) * p^k * (1-p)^(n-k), 여기서 (n C k)는 조합을 나타낸다.
