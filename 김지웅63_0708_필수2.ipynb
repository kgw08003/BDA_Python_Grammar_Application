```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

%matplotlib inline
```

```python
df = pd.read_csv('scores400.csv')
```

```python
df
```

```python
sample = np.random.choice(df['score'],50) #50개 추출
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.hist(sample, bins=100, range=(1,100), density= True)
ax.set_xlim(0,100)
ax.set_ylim(0, 0.05)
ax.set_xlabel('score')
plt.show()
```
