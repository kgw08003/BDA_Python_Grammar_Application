# 필수과제 2
- 정규성 검정, 표본에 따라서 달라지는 분석기법에 대해서 리마인드 공부
- 정리해 주세요, 분산분석까지도 개념 정리해 주시고 복습해서 다음시간까지 제출해주세요!

### 정규성 검정
- 정규성 검정은 주어진 데이터가 정규 분포를 따르는지를 확인하기 위한 통계적 분석 방법
- 데이터가 정규 분포를 따를때, 일부 통계적 분석 방법들은 이를 가정으로 정확한 결과를 도출할 수 있다.
- 정규성 검정은 표본의 수에 따라서 검정에 사용되는 분석기법이 달라질 수 있다.

### 표본이 1개인 경우
#### 모평균에 대한 유의수준 단측검정
- 주어진 표본 데이터를 사용하여 모집단의 평균이 어떤 특정한 값보다 크거나 작은지를 확인하는 통계적 검정 방법
- 모집단의 평균과 평균 간의 차이가 한쪽 방향으로 크게 존재하는지를 확인하기 위해서 사용되는 검정방법
- 유의수준을 한 쪽 꼬리에 할당하고, 해당 방향으로의 p-값과 비교하여 가설을 검정
-----------


#### 모평균에 대해 검정하는데 모분산을 모르는 경우:
 - t-분포를 이용하여 검정을 수행. 표본 크기가 크면(일반적으로 30 이상) 중심극한정리에 의해 표본 평균의 분포가 정규분포에 근사하므로, t-분포 대신에 z-분포(표준정규분포)를 사용할 수도 있음.

#### 모분산에 대한 검정:
- 주어진 데이터로부터 얻은 표본 분산과 모집단의 분산 간의 차이가 우연에 의한 것인지를 판단하는 검정 방법. 주로 카이제곱 분포를 이용하여 가설을 검정.
- 표본을 사용하여 표본 분산을 추정하고, 추정된 표본 분산을 이용하여 검정 통계량을 계산합니다. 계산된 검정 통계량을 자유도와 함께 카이제곱 분포의 임계값과 비교하여 가설을 검정

### 표본이 2개인 경우:
- 표본이 2개인 경우에는 주로 등분산성 검정과 t-검정이 함께 사용될 수 있으며 같은 개체 또는 그룹에서 두 가지 조건 처리를 비교하는 경우에 사용
- 등분산성 검정은 두 표본이 같은 분산을 가지는지를 확인하기 위한 검정 방법이다.
- 만약 두 표본이 등분산성을 만족한다면, t-검정을 통해 두 표본 간의 평균 차이가 유의한지를 판단할 수 있다. 
- 등분산성을 만족하지 않는 경우에는 등분산성 가정을 위배한 t-검정인 Welch's t-test를 사용할 수 있다.
- Shapiro-Wilk 검정은 표본의 크기가 작을 때(일반적으로 50개 이하) 정확한 결과를 제공하는 검정 방법으로 검정은 표본 데이터가 정규 분포를 따르는지를 확인하는데 사용된다.
------------
## 대응표본일때- 같은 개체 또는 그룹에서 두 가지 조건 또는 처리를 비교하는 경우에 사용
##### 정규분포를 가정할 수 있을 때: 대응비교 t-검정 (Paired t-test)
- 데이터에서 개체 또는 그룹의 차이 값을 계산(두 조건의 관찰값 차이)
- 차이 값의 평균과 표준편차를 계산
- t-검정 통계량(평균 차이를 표준오차로 나눈 값)을 계산하여 우연에 의한 차이 판단
- t-검정 통계량을 유의수준(일반적으로 0.05)에 대한 임계값과 비교하여 가설 검정

##### 정규분포를 가정할 수 없을 때: 윌콕슨 부호순위검정 (Wilcoxon Signed-Rank Test)
- 데이터에서 개체 또는 그룹의 차이 값을 계산(두 조건의 관찰값 차이)
- 차이 값의 절댓값을 기준으로 순위를 매김(같은 값은 같은 순위, 작은 값부터 시작)
- 양수 또는 음수인 순위 합계를 계산(양수 순위의 합과 음수 순위의 합 중 작은 값 선택)
- 순위 합계를 유의수준(일반적으로 0.05)에 대한 임계값과 비교하여 가설 검정
-------------------
## 독립표본일때 - 서로 다른 개체 또는 그룹에서 각각의 처리 또는 조건을 비교하는 경우
##### 정규분포를 가정할 수 있을 때: 대응비교 t-검정 (Paired t-test)을 사용한다.
- 각 개체 또는 그룹의 관찰값 평균과 표준편차를 계산합니다.
- 두 개체 또는 그룹의 평균 차이를 계산합니다.
- 가설 검정을 위해 t-검정 통계량을 계산합니다. 이 통계량은 평균 차이를 표준오차로 나눈 값을 사용해 두 개체 또는 그룹의 차이가 우연에 의한 것인지 판단합니다.
- 계산된 t-검정 통계량을 유의수준(보통 0.05)의 임계값과 비교해 가설을 검정합니다.
##### 정규분포를 가정할 수 없는 경우, 만위트니 U 검정(Mann-Whitney U Test)을 사용한다.
- 각 개체 또는 그룹의 관찰값을 합친 후 순위를 매깁니다. 같은 값은 평균 순위를 부여하며, 작은 값부터 시작해 순차적으로 증가합니다.
- 두 개체 또는 그룹의 순위 합을 계산합니다. 이 합은 일반적으로 작은 값으로 선택됩니다.
- 계산된 순위 합을 유의수준(보통 0.05)의 임계값과 비교해 가설을 검정합니다.
- 독립 표본은 서로 다른 개체나 그룹 간 처리 또는 조건의 비교에 사용되며, 데이터의 정규성 여부에 따라 독립 비교 t-검정 또는 만위트니 U 검정 중 하나를 선택하여 사용할 수 있습니다.

### 표본이 3개 이상인 경우:
- 표본이 3개 이상인 경우에는 주로 분산 분석(ANOVA)이 사용된다.
- 분산 분석은 여러 개의 표본 간에 평균 차이가 있는지를 검정하는 방법 
- 일반적으로 등분산성을 가정한다.
- 등분산성을 가정한 경우에는 일원분산분석(One-way ANOVA)을 사용하고, 등분산성 가정을 위배하는 경우에는 Welch의 분산 분석(Welch's ANOVA)이나 Kruskal-Wallis 검정 등을 사용할 수 있게 된다.
