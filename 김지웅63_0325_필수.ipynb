# 필수과제

- 타이타닉데이터로 수업시간에 머신러닝을 진행할건데 여러분들이 전처리한 데이터를 가지고 진행할 것
- 여러분들이 전처리한 데이터에 대한 정확도나 다른 지표들을 확인할 수 있다.
- 전처리는 모두 동일하게 하지 않을 것이기 때문에 실제 다른 분들의 전처리를 보고 그걸 통해 나온 정확도 등을 확인해보며 어떤지 알아보는 시간을 가진다.

- 타이타닉 데이터를 전처리 꼭 진행 주시고 하나의 데이터프레임으로 만들어 주시는데
- 문자열 데이터가 다 인코딩이 되고, 수치형 데이터도 결측치가 없는 하나의 깔끔한 데이터 프레임으로 만들어 제출해 주셔야 합니다.
- 파일 형식은 csv 파일입니다.
- 가장 우선적으로는 조별로 진행해 주시는 분들 데이터를 가지고 수업시간 확인할 것
- 부족하다면 개인의 과제를 랜덤하게 추출할 예정

- csv파일만 제출하는 게 아니라 구체적인 코드와 주석과 왜 그렇게 하셨는지? 왜 전처리를 그렇게 했는지 정확하게 설명해 주셔야 합니다.


## - 타이타닉 데이터 설명

1. survived 생존여부(0 = 사망, 1 = 생존)
2. pclass 객실의 등급 (1 = 1st, 2 = 2nd, 3 = 3rd )
3. sex 성별
4. age 연령
5. sibsp 타이타닉호에 탑승한 형제 or 배우자의 수
6. parch 타이타닉호에 탑승한 부모 or 자녀의 수
7. fare 승객요금
8. embarked 기항지 위치 (S, Q, C)
9. class 객실의 등급
10. who 남성, 여성, 아동
11. adult_male 성인남자인지 (true면 성인남자)
12. deck 선실번호 알파벳 (A,B,C,D,E,F,G)
13. embark_town 탑승지 명칭
14. alive 생존여부
15. alone 혼자 탑승했는지

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.info()
```

```python
df.isna().sum() # 결측치 확인
```

```python
df_1 = df.copy() # copy()를 통해 혹시모를 데이터 변질 방지
```

- 결측치 ( 결측치 처리 필요 데이터)  : age, embarked, deck, embark_town

- 문자형 ( 수치형 인코딩 필요 데이터) : sex, embarked, who, embark_town, alive

```python
df_1 = df.dropna(subset=['embark_town']) # 결측치가 2개 이지만 문자형이고, 값을 알지 못하니까 드롭을 통해서 결측치 제거
```

```python
df_1
```

```python
age_mean = df.age.mean() # age의 평균값을 측정
```

```python
age_mean
```

```python
df_1['age'] = df_1['age'].fillna(age_mean) # age는 평균값으로 그 결측치로 대처함
df_1
```

```python
df_1.isna().sum() # 중간 결측치 확인
```

```python
df_1 = df_1.drop(['deck'], axis=1) # 결측치인 값이 70% 비율을 넘어감으로 삭제처리함
df_1
```

```python
df_1.isna().sum() # 최종 결측치 확인
```

## 이를통해 결측치값이 0로 다 세팅 된 것을 확인하였다.

### 원 핫 인코딩을 통해 문자열을 숫자형으로 변환하는 작업 시작
- 문자형 ( 수치형 인코딩 필요 데이터) : sex, embarked, who, embark_town, alive

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_1.info()
```

```python
df_1 = pd.get_dummies(df_1, columns = ['sex'], drop_first=True)
df_1
```

```python
df_1.embarked.unique() #embarekd는 3개로 구성됨
```

```python
# embarked (범주형 데이터로 변환)
df_1 = pd.get_dummies(df_1, columns = ['embarked'])
df_1
```

```python
df_1.who.unique() # who는 3개로 구성됨
```

```python
# who (범주형 데이터로 변환)
df_1 = pd.get_dummies(df_1, columns = ['who'])
df_1
```

```python
df_1.embark_town.unique()
```

```python
# embark_town (범주형 데이터로 변환)
df_1 = pd.get_dummies(df_1, columns = ['embark_town'])
df_1
```

```python
df_1.alive.unique()
```

```python
# alive → alive_yes(범주형 데이터로 변환) 1 이면 생존 0이면 죽음
df_1 = pd.get_dummies(df_1, columns = ['alive'], drop_first=True)
df_1
```

```python
df_1.info()
```

```python
#pclass는 정수형 데이터 이지만 1~3등급을 의미하는 범주형 데이터로 보는 것이 더 빠른 머신학습을 유도할 수 있다.
df_1 = pd.get_dummies(df_1, columns = ['pclass'])
df_1
```

```python
df_1 = pd.get_dummies(df_1, columns = ['adult_male']) #adult_male 불함수 데이터 이지만 TRUE of FALSE를 의미하므로, 범주형으로 바꾸어 보았다.
df_1
```

```python
df_1 = pd.get_dummies(df_1, columns = ['class']) #class는 category 이지만 First, Second, Third로 존재함으로 범주형 데이터로 바꿔보았다.
df_1
```

```python
df_1 = pd.get_dummies(df_1, columns = ['alone']) #aloned은 불함수 데이터 이지만 TRUE of FALSE를 의미하므로, 범주형으로 바꾸어 보았다.
df_1
```
