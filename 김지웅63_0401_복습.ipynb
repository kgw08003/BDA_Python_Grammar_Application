```python
df1 = df1.select_dtypes(include = [np.number]) #수치형 데이터만 추출하기 
```

```python
df1=df1[['survived', 'age', 'sibsp', 'parch', 'fare', 'sex_male',
       'embarked_C', 'embarked_Q', 'embarked_S', 'who_child', 'who_man',
       'who_woman', 'embark_town_Cherbourg', 'embark_town_Queenstown',
       'embark_town_Southampton', 'alive_yes', 'pclass_1', 'pclass_2',
       'pclass_3', 'adult_male_False', 'adult_male_True', 'class_First',
       'class_Second', 'class_Third', 'alone_False', 'alone_True']]
```

```python
df1
```

```python
x_df1 = df1.drop('survived',axis=1)
y_df1 = df1['survived']
```

```python
x_df1
```

```python
y_df1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_df1, y_df1, test_size=0.2, random_state=111)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# 결정트리, 랜포, 로지스틱 회귀 만들기
dt_clf = DecisionTreeClassifier(random_state=111)
rf_clf = RandomForestClassifier(random_state=111)
lr_clf = LogisticRegression(solver ='liblinear')
```

```python
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DT정확도:{0:.4f}'.format(accuracy_score(y_test, dt_pred)))

rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
print('RF정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))

lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
print('LR정확도:{0:.4f}'.format(accuracy_score(y_test, lr_pred)))
```
