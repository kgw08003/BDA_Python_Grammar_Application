```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df= pd.read_csv('wine_test.csv')
```

```python
df.info()
```

```python
df['alcohol']=df['alcohol'].fillna(df['alcohol'].median())
df['nonflavanoid_phenols']=df['nonflavanoid_phenols'].fillna(df['nonflavanoid_phenols'].mean())
```

```python
df.columns
```

```python
df_n=df[['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash',
       'magnesium', 'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
       'proanthocyanins', 'color_intensity', 'hue',
       'od280/od315_of_diluted_wines', 'proline', 'class']]
```

```python
# 훈련데이터와 테스트데이터 나누기
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_n.drop('class',axis=1), df_n['class'],
                                                   test_size=0.3, random_state=111)
```

```python
# 스케일링 기법
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
st_scaler= StandardScaler()
mm_scaler=MinMaxScaler()
rb_scaler=RobustScaler()
```

```python
## knn 모델
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
# 정밀도
from sklearn.metrics import accuracy_score
```

**Standard Scaler:

```python
# Standard Scaler 초기화
st_scaler = StandardScaler()

# 훈련 데이터에 스케일링 적용
x_train_st = st_scaler.fit_transform(x_train)
x_test_st = st_scaler.transform(x_test)

# 스케일링 된 데이터 출력
print("Scaled Training Data with Standard Scaler:\n", x_train_st)
print("\nScaled Testing Data with Standard Scaler:\n", x_test_st)
```

**MinMax Scaler:

```python
# MinMax Scaler 초기화
mm_scaler = MinMaxScaler()

# 훈련 데이터에 스케일링 적용
x_train_mm = mm_scaler.fit_transform(x_train)
x_test_mm = mm_scaler.transform(x_test)

# 스케일링 된 데이터 출력
print("Scaled Training Data with MinMax Scaler:\n", x_train_mm)
print("\nScaled Testing Data with MinMax Scaler:\n", x_test_mm)
```

**Robust Scaler:

```python
# Robust Scaler 초기화
rb_scaler = RobustScaler()

# 훈련 데이터에 스케일링 적용
x_train_rb = rb_scaler.fit_transform(x_train)
x_test_rb = rb_scaler.transform(x_test)

# 스케일링 된 데이터 출력
print("Scaled Training Data with Robust Scaler:\n", x_train_rb)
print("\nScaled Testing Data with Robust Scaler:\n", x_test_rb)
```
