### 베르누이 분포 (Bernoulli Distribution):

- 베르누이 분포는 이항 분포의 특수한 경우로, 단 하나의 이진 결과를 가지는 확률 변수의 분포입니다.
- 확률 변수의 값이 1일 확률을 p라고 할 때, 0일 확률은 1-p입니다.

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.3  # 확률 p
bernoulli_data = np.random.binomial(1, p, 1000)  # 1000개의 베르누이 분포 샘플 생성
unique, counts = np.unique(bernoulli_data, return_counts=True)
frequencies = counts / len(bernoulli_data)

plt.bar(unique, frequencies)
plt.xticks([0, 1], ['0', '1'])
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution')
plt.show()
```

- 확률 p가 0.3인 베르누이 분포에서 1000개의 샘플을 생성하고, 각 값의 빈도를 계산하여 시각화함
- 그래프에서 x축의 rkqt(0과1)을 나타내고, y축은 확률을 나타낸다.
- 베르누이 분포의 경우, 0과 1의 확률이 각각 0.7과 0.3으로 주어진 확률 p에 따라 달라진다.

### 이항 분포
- 이항 분포는 독립적인 베르누이 시행을 n번 반복한 후 성공(1)의 횟수를 나타내는 확률 변수의 분포입니다.
- 각 베르누이 시행은 확률 p로 성공하거나 1-p로 실패하는 두 가지 결과를 갖습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10  # 시행 횟수
p = 0.3  # 성공 확률 p
binomial_data = np.random.binomial(n, p, 1000)  # 1000개의 이항 분포 샘플 생성
unique, counts = np.unique(binomial_data, return_counts=True)
frequencies = counts / len(binomial_data)

plt.bar(unique, frequencies)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.show()
```

### 기하분포
- 기하 분포는 베르누이 시행에서 처음 성공할 때까지의 실패 횟수를 나타내는 확률 변수의 분포입니다.
- 각 베르누이 시행은 확률 p로 성공하거나 1-p로 실패하는 두 가지 결과를 갖습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.3  # 성공 확률 p
geometric_data = np.random.geometric(p, 1000)  # 1000개의 기하 분포 샘플 생성
unique, counts = np.unique(geometric_data, return_counts=True)
frequencies = counts / len(geometric_data)

plt.bar(unique, frequencies)
plt.xlabel('Number of Failures before First Success')
plt.ylabel('Probability')
plt.title('Geometric Distribution')
plt.show()
```

- 성공 확률 p가 0.3인 기하 분포에서 1000개의 샘플을 생성하고, 처음 성공하기 까지의 실패 횟수의 빈도를 계산하여 시각화
- 그래프 x축은 처음 성공하기 까지의 실패 횟수, y축은 확률을 나타낸다.
- 기하분포는 성공확률 p에 따라 분포의 모양이 달라질 수 있다.

### 포아송 분포
- 포아송 분포는 단위 시간 또는 공간에서 발생하는 사건의 수를 나타내는 이산형 확률 분포입니다.
- 사건이 발생하는 평균 비율인 λ에 따라 분포의 형태가 결정됩니다.

```python
from scipy.stats import poisson

x = np.arange(41, dtype='f8') #큰 숫자 취급을 위해 데이터 타입을 실수로 변경
l=20
y = poisson(l).pmf(x)

plt.figure(figsize=(10, 8))
plt.plot(x, y, 'o--')
plt.xlabel('X')
plt.ylabel('P(X)')
plt.show()
```

### 정규 분포
- 정규 분포는 가장 잘 알려진 연속형 확률분포로, 자연 현상에서 많이 관찰되는 데이터의 분포를 근사하는 데 사용됩니다.
- 평균(mu)과 표준 편차(sigma)라는 두 개의 매개변수로 정의됩니다.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

x = np.arange(0, 20, 0.03) # 범위 0에서 20까지
plt.figure(figsize=(10,8))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, norm.pdf(x, loc=10, scale=2)) #평균이 10이고 표준편차가2인 정규분포
plt.show()
```

### 카이제곱 분포
- 카이제곱 분포는 분산을 추정하거나 독립성을 검정하는 등 다양한 통계적 분석에 사용되는 연속형 확률분포입니다.
- 자유도(degree of freedom)라는 매개변수에 의해 정의됩니다.

```python
import numpy as np
import matplotlib.pyplot as plt

df = 3  # 자유도
chi_squared_data = np.random.chisquare(df, 1000)  # 1000개의 카이제곱 분포 샘플 생성

plt.hist(chi_squared_data, bins=30, density=True)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution')
plt.show()
```

```python
import scipy.stats as Chi2

chi2=np.linspace(0.5, 50, 100)
y = Chi2.chi2(1).pdf(chi2)        

plt.figure(figsize=(10, 8)) # 플롯 사이즈 지정
plt.plot(chi2, y, color="blue", label=r'$\chi^2$(1)')         
plt.xlabel(r'$\chi^2$')
plt.ylabel("y") 
plt.show()  
```

### 지수 분포
- 지수 분포는 사건이 발생하는 간격을 모델링하는 데 사용되는 연속형 확률분포입니다.
- 사건 발생률(parameter)인 람다(lambda)에 의해 정의됩니다.

```python
from scipy.stats import expon
x = np.linspace(0, 10, 101)
loc=0
l=1
y=expon(loc,l).pdf(x)

plt.figure(figsize=(10, 8))
plt.plot(x, y, 'b')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

- t 분포
- t 분포는 표본 평균의 분포를 모델링하는 데 사용되는 연속형 확률분포입니다.
- 자유도(degree of freedom)라는 매개변수에 의해 정의됩니다.
- 모집단 표준편차를 알 수 없을 때 표본 평균과 모집단 평균 사이 표준화된 거리이며 규 분포의 평균을 측정할 때 주로 사용되는 분포로자유도가 클수록 표준정규분포에 가깝다

```python
import scipy.stats as T

t=np.linspace(-5,5,51) #-5부터 5까지 50등분
y = T.t(5).pdf(t)        

plt.figure(figsize=(10, 8))
plt.plot(t, y, color="b")     
plt.xlabel("t") 
plt.ylabel("y") 
plt.legend(["t(10)"])
plt.show() 
```

### F분포
- F 분포는 분산 비율을 비교하는 분석에 사용되는 연속형 확률분포입니다.
- 두 개의 자유도 매개변수(df1, df2)로 정의됩니다.

```python
import scipy.stats as stats

x = np.linspace(0,5,201) #0부터 5까지 200등분

nu1 = 4  #자유도1 - 4
nu2 = 40 #자유도2 - 40

y=stats.f(nu1,nu2).pdf(x)

plt.figure(figsize=(10, 8))
plt.plot(x, y, color="b", label='F(4, 40)')
plt.xlabel('F')
plt.ylabel("y")
plt.show()
```
