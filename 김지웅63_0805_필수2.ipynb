```python
# 결측치 합  
df.isna().sum()
```

```python
# 결측치가 많이없으니까 drop, 중위값, 평균값 으로 하자
df['alcohol'].fillna(df['alcohol'].mean(), inplace=True)   #알코올 피처 결측치를 열 평균값으로 대체

# 'nonflavanoid_phenols' 열의 결측치를 열의 평균값으로 대체
mean_value = df['nonflavanoid_phenols'].mean()
df['nonflavanoid_phenols'].fillna(mean_value, inplace=True)
```
