#### 머신러닝(ML)이란?
- 머신(컴퓨터)에게 학습을 시켜서 문제를 해결하는 것
---
정답을 아는지 여부에 따라 학습 종류가 분류된다
1. 지도 학습 : 정답을 알고, 정답을 기반으로 학습을 하는 것
2. 비지도 학습 : 정답 없이 머신을 학습시켜 인간이 생각하지 못한 패턴 등을 분석하여 스스로 정답을 찾도록 하는 것
3. 준지도 학습 : 지도 학습과 비지도 학습이 섞여있는 것
4. 강화 학습 : 보상에 대한 개념으로, 가장 큰 보상을 얻기 위해 최상의 전략을 스스로 학습하는 것(가중치 고려)
5. 딥러닝 : 뉴럴 네트워크와 같은 원리로 인공 신경망을 만들어서 학습하는 것

#### 배울 내용
- 머신러닝 - 지도학습
- 분류와 회귀
---
- 분류 : 이산형으로 범주형(카테고리)의 개념이다, 연속적인 숫자가 아닌 정답 레이블이 0, 1 특정한 값으로 이루어진다.
- 회귀 : 연속형으로, 정답 데이터가 연속적인 값이다.
---
즉, 정답 데이터의 타입에 따라 나뉜다

#### 머신러닝 지도학습, 분류
이진 분류/ 다중 분류

---
1. 0,1 / 스팸 or 스팸x / 고객 or 고객x 등 -> 이진 분류
2. 0,1,2 등 분류할 값이 다중인 경우 -> 다중 분류

#### 데이터 전처리 : 머신러닝에서 가장 중요한 작업 중 하나
- 전처리 : 데이터를 변환하는 작업 처리를 하는 것
- 머신은 숫자로만 학습을 하기 때문에 머신러닝을 하기 위해서는 모든 데이터를 숫자로 바꿔주어야 한다.(자연어나 영어, 문자열 데이터 등은 머신이 이해를 하지 못한다)
#### 즉, 머신러닝을 위해 전처리를 한다는 것은 머신에게 학습 시키기 위한 데이터프레임을 만드는 것이다(데이터프레임을 만드는 것은 결국 머신이 학습할 수 있도록 수치형 데이터로 변환하는 것이다)

#### 머신러닝 : 성능이 좋아진다
- 머신러닝이란 과거 데이터를 기반으로 실제 예측을 하는 것이다
- 과거 데이터를 기반으로 실제를 예측해보고, 정답과 비교해서 과거 데이터의 정답과 가장 유사하면 예측을 잘하는 것이라고 본다

#### 정답 데이터가 가장 중요한 키워드이다
- 기존에 나와 있는 데이터를 train과 test로 나눈다
- train과 test로 나누지 않고 전체 데이터로 학습을 하면 과대적합이 발생한다

---
따라서 train을 가지고 학습한 후 실제 예측을 하는데 그 예측값을 test 데이터와 비교하여 모델 성능을 평가한다
