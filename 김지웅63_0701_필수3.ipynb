#### 상관계수
- 상관계수는 두 변수 간의 선형적인 관계의 강도와 방향을 나타내는 통계적 지표
- 상관계수는 -1과 1사이의 값을 가지며, 다음과 같은 특징을 가진다.


- 상관계수가 1에 가까울수록 양의 선형 관계가 강하다. 즉, 한 변수가 증가하면 다른 변수도 증가하는 경향을 보인다.
- 상관계수가 -1에 가까울수록 음의 선형 관계가 강하다. 즉, 한 변수가 증가하면 다른 변수는 감소하는 경향을 보인다.
- 상관계수가 0에 가까울수록 두 변수 간에는 선형적인 관계가 없다. 하지만 다른 종류의 관계가 있을 수 있으므로 주의가 필요하다.

##### 상관계수는 주어진 데이터 집합에서 두 변수 X와 Y 간의 관계를 측정하기 위해 사용됩니다. 상관계수는 다음과 같이 계산된다.

###### 상관계수(Correlation(X, Y)) = Cov(X, Y) / (σX * σY)

- Correlation(X, Y): X와 Y의 상관계수
- Cov(X, Y): X와 Y의 공분산
- σX, σY: X와 Y의 표준편차

- 1에 가까울수록 양의 선형 관계가 강해지고, 값이 -1에 가까울수록 음의 선형 관계가 강해진다.
