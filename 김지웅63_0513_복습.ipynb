## 반복문을 통한 데이터 프레임 접근하기
- 반복문에 들어가는 값
- 반복문을 통하여 나오는 값

```python
for i in range(5):
    print(i)
```

- range는 range(5)라는 숫자가 나옴.
- in 뒤에는 range함수와 시퀀스 객체에 따라 값이 달라진다.

```python
a = ['버섯','소고기','돼지고기','김치','쌀']
```

```python
for i in a:
    print(i)
```

- 시퀀스 객체 등으로 range가 아닌 다른 것으로 끝나게 될 경우, 시퀀스 객체 안에 값을 가지고 오게 된다.
- 시퀀스 객체를 range()를 이용하여 표현하기

```python
for i in range(len(a)):
    print(a[i])
```

- 시리즈 형태로 반복문 이용해 조건물을 더욱 정교하게 만들자
- mpg 데이터 중에서 trans 변수에 s6 값 있는 데이터를 출력하여 제조사별 카운팅
- l5만 추출하는 함수를 만들어보자

```python
mpg
```

```python
def get_trans_spe(series):
    tr_hh = len([_ for x in series if x[0] in 'm5'])
    return tr_hh
```

- 제조사별 카운팅해야함

```python
get_trans_spe(mpg.trans)
```

```python
mpg.groupby('manufacturer').agg({'trans':get_trans_spe}).reset_index()
```

- 연도별로 cty 평균이 16미만 친구들과 16이상이 친구들을 묶어보자!

```python
def get_y_cty(series):
    if mpg.iloc[series.index[0], : ]['year']==2008:
        cty_cnt = len([_ for x in series if x>=16])
    else:
        cty_cnt = len([_ for x in series if x<16])
    return cty_cnt
```

```python
mpg.groupby('year').agg({'cty':get_y_cty}).reset_index()
```
